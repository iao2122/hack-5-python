```python
import numpy as np
import pandas as pd
import requests
```

```python
url2 = "http://eaton-lab.org/data/iris-data-dirty.csv"
with open("./iris-data-dirty.csv", 'w') as ffile:
    ffile.write(requests.get(url2).text)
df = pd.read_csv("iris-data-dirty.csv", names=['sl', 'sw', 'pl', 'pw', 'species'])
```

```python
df = df.replace("Iris-setsa", "Iris-setosa")
df = df.replace("Iris-versicolour", "Iris-versicolor")
df
```

```python
gb = df.groupby('species')
for id, d in gb:
    print(id)
    display(d.describe())
```
