```python
import numpy as np
import pandas as pd
import requests

from scipy.stats import ttest_ind
```

```python
url2 = "http://eaton-lab.org/data/iris-data-dirty.csv"
with open("./iris-data-dirty.csv", 'w') as ffile:
    ffile.write(requests.get(url2).text)
df = pd.read_csv("iris-data-dirty.csv", names=['sl', 'sw', 'pl', 'pw', 'species'])
```

```python
df = df.replace("Iris-setsa", "Iris-setosa")
df = df.replace("Iris-versicolour", "Iris-versicolor")
df = df.dropna()
df.head()
```

```python
df.describe()
```

```python
gb = df.groupby('species')
for id, d in gb:
    print(id)
    display(d.describe())
```

```python
gb.get_group("Iris-setosa").head()
```

```python
df_setosa = gb.get_group("Iris-setosa")
df_versicolor = gb.get_group("Iris-versicolor")
```

```python
ttest_ind(df_setosa["sl"], df_versicolor["sl"])
```
